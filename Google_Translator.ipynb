```python
from tkinter import*
from tkinter import ttk
from googletrans import Translator,LANGUAGES

def change(text='type',src='English',des='Urdu'):
    text1=text
    src1=src
    des1=des
    trans=Translator()
    trans1=trans.translate(text,src=src1,des1=des)
    return trans1.text
def final():
    s=com_sor.get()
    d=com_des.get()
    msg=Sor_txt.get(1.0,END)
    text_1=change(text=msg,src=s,des=d)
    des_txt.delete(1.0,END)
    des_txt.insert(END,text_1)
    

root=Tk()
root.title("Translator")
root.geometry("500x600")
root.config(bg='brown')
lab_txt=Label(root,text='Translator',font=("Time New Roman",40,"bold"))
lab_txt.place(x=100,y=40,height=50,width=300)

frame=Frame(root).pack(side=BOTTOM)

lab_txt=Label(root,text='Source Text',font=("Time New Roman",20,"bold"))
lab_txt.place(x=100,y=100,height=20,width=300)

Sor_txt=Text(frame,font=('Time New Roman',10),wrap=WORD)
Sor_txt.place(x=60,y=125,height=150,width=380)

list_text=list(LANGUAGES.values())

com_sor=ttk.Combobox(frame,value=list_text)
com_sor.place(x=60,y=290,height=25,width=100)
com_sor.set('English')

set_boton=Button(frame,text="Translate",relief=RAISED,command=final)
set_boton.place(x=200,y=290,height=25,width=100)

com_des=ttk.Combobox(frame,value=list_text)
com_des.place(x=340,y=290,height=25,width=100)
com_des.set('Urdu')

lab_txt=Label(root,text='Translation',font=("Time New Roman",20,"bold"))
lab_txt.place(x=100,y=335,height=20,width=300)

des_txt=Text(frame,font=('Time New Roman',10),wrap=WORD)
des_txt.place(x=60,y=365,height=180,width=380)


root.mainloop()
```
